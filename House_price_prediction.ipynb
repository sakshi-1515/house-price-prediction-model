```python
# function to convert ranges of total sq feets to normal numer
def convert_sqft(x):
    if "-" in x:
        parts = x.split("-")
        return (float(parts[0]) + float(parts[1])) / 2
    try:
        return float(x)
    except:
        return None
```

```python
df1["total_sqft"] = df1["total_sqft"].apply(convert_sqft)
```
